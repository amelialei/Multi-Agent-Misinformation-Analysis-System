```python
base_df = pd.read_csv("base_outputs.csv")
cot_df = pd.read_csv("cot_outputs.csv")
fcot_df = pd.read_csv("fcot_outputs.csv")
```

```python
score_cols = ['freq_score', 'mal_score', 'sens_score', 'naive_score', 'overall_confidence']

base_df['technique'] = 'Base'
cot_df['technique'] = 'CoT'
fcot_df['technique'] = 'Fractal CoT'

base_vis = base_df[['url'] + score_cols + ['technique']]
cot_vis = cot_df[['url'] + score_cols + ['technique']]
fcot_vis = fcot_df[['url'] + score_cols + ['technique']]

combined_df = pd.concat([base_vis, cot_vis, fcot_vis], ignore_index=True)

melted_df = combined_df.melt(id_vars=['url','technique'],
                             value_vars=score_cols,
                             var_name='score_type',
                             value_name='score')
```

```python
melted_df.head()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=melted_df, x='score_type', y='score', hue='technique')
plt.title('Comparison of Scoring Across Prompting Techniques')
plt.ylabel('Score')
plt.xlabel('Score Type')
plt.legend(title='Prompting Technique')
plt.show()
```

```python
factor_melted = melted_df[melted_df['score_type'] != 'overall_confidence']

plt.figure(figsize=(12,6))
sns.boxplot(data=factor_melted, x='score_type', y='score', hue='technique')
plt.title('Factor Scores Comparison Across Prompting Techniques')
plt.ylabel('Score (0-2)')
plt.xlabel('Score Type')
plt.ylim(0,3)
plt.legend(title='Prompting Technique')
plt.show()
```

```python
confidence_melted = melted_df[melted_df['score_type'] == 'overall_confidence']

plt.figure(figsize=(8,5))
sns.boxplot(data=confidence_melted, x='technique', y='score')
plt.title("Overall Confidence Comparison Across Prompting Techniques")
plt.ylabel("Confidence (%)")
plt.xlabel("Prompting Technique")
plt.ylim(80,120)
plt.show()
```

```python
avg_scores = melted_df.groupby(['technique', 'score_type']).score.mean().reset_index()
avg_scores
```

```python
diff_df = fcot_vis.set_index('url')[score_cols] - base_vis.set_index('url')[score_cols]

plt.figure(figsize=(12,6))
sns.heatmap(diff_df, annot=True, cmap='coolwarm', center=0)
plt.title('Score Differences per Article (Fractal CoT - Base)')
plt.ylabel('Articles')
plt.xlabel('Score Type')
plt.show()
```

```python
factor_cols = ['freq_score', 'mal_score', 'sens_score', 'naive_score']

diff_factors = fcot_vis.set_index('url')[factor_cols] - base_vis.set_index('url')[factor_cols]

plt.figure(figsize=(12,6))
sns.heatmap(diff_factors, annot=True, cmap='coolwarm', center=0)
plt.title('Factor Score Differences per Article (Fractal CoT - Base)')
plt.ylabel('Articles')
plt.xlabel('Factor')
plt.show()
```

```python
diff_conf = fcot_vis.set_index('url')['overall_confidence'] - base_vis.set_index('url')['overall_confidence']

plt.figure(figsize=(8,6))
sns.heatmap(diff_conf.to_frame(), annot=True, cmap='coolwarm', center=0)
plt.title('Overall Confidence Differences per Article (Fractal CoT - Base)')
plt.ylabel('Articles')
plt.xlabel('Overall Confidence')
plt.show()
```

```python
def normalize_score(row):
    if row['score_type'] == 'overall_confidence':
        return row['score'] / 100
    else:
        return row['score'] / 2

avg_scores['score_norm'] = avg_scores.apply(normalize_score, axis=1)

# Radar plot setup
labels = avg_scores['score_type'].unique()
num_labels = len(labels)
angles = np.linspace(0, 2*np.pi, num_labels, endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(8,8))
for tech in avg_scores['technique'].unique():
    scores = avg_scores[avg_scores['technique']==tech]['score_norm'].tolist()
    scores += scores[:1]
    plt.polar(angles, scores, label=tech, marker='o', linewidth=2)

plt.xticks(angles[:-1], labels)
plt.title("Normalized Score Profiles Across Prompting Techniques")
plt.legend(loc='upper right')
plt.show()
```
